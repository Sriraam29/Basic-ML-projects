```python
selected_columns = [
    # Demographics
    'Country',
    'Age',
    'Gender',
    'EdLevel',
    'UndergradMajor',
    # Programming experience
    'Hobbyist',
    'Age1stCode',
    'YearsCode',
    'YearsCodePro',
    'LanguageWorkedWith',
    'LanguageDesireNextYear',
    'NEWLearn',
    'NEWStuck',
    # Employment
    'Employment',
    'DevType',
    'WorkWeekHrs',
    'JobSat',
    'JobFactors',
    'NEWOvertime',
    'NEWEdImpt'
]
```

```python
survey['Age1stCode']=pd.to_numeric(survey.Age1stCode,errors='coerce')
```

```python

survey['YearsCode'] = pd.to_numeric(survey.YearsCode, errors='coerce')
survey['YearsCodePro'] = pd.to_numeric(survey.YearsCodePro, errors='coerce')
```

```python
survey.info()
```

```python
survey.describe()
```

```python
survey.drop(survey[survey.Age<10].index,inplace=True)
```

```python
survey.drop(survey[survey.WorkWeekHrs>70].index,inplace=True)
survey.drop(survey[survey.Age>80].index,inplace=True)
```

```python
survey.describe()
```

```python
survey['Gender'].value_counts()
```

```python
survey.drop(~survey[survey.Gender  not in['Man','Women']].index,inplace=True)
```

```python
survey.drop(survey[~survey.Gender.isin(['Man', 'Woman'])].index, inplace=True)
```

```python
survey.drop(survey[survey.isna().index()==True])
```
